
note：$ceil(2n/3)=\left\lceil {2n/3} \right\rceil$；$:=$ 表示赋值, 等价于 $\to$；A是一个包含n的整数元素的数组, 

1. 写出该算法时间复杂度的递归公式, 并求解其对应的渐进表示
2. 描述一下该算法的功能, 并判断是否是最高效的解决方案
3. \* 使用python编程实现上述算法或其对应的更高效版本的算法

idea:

````python
import math

def DoSomething(A, p, r):
    n = r - p + 1
    if n == 2:
        if A[p] > A[r]:
            A[p], A[r] = A[r], A[p]
    elif n >= 3:
        m = math.ceil(2 * n / 3)
        DoSomething(A, p, p + m - 1)
        DoSomething(A, r - m + 1, r)
        DoSomething(A, p, p + m - 1)

# 测试
A = [9, 7, 5, 3, 1, 2, 4, 6, 8]
DoSomething(A, 0, len(A)-1)
print("结果:", A)
# your algorithm time complexity is: O(n^log3(2))
````

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9

idea:

````python
def maxProfit(profits):
    max_sum = float('-inf')
    current_sum = 0
    start = 0
    buy_day = sell_day = 0
    temp_start = 0

    for i, p in enumerate(profits):
        if current_sum <= 0:
            current_sum = p
            temp_start = i
        else:
            current_sum += p

        if current_sum > max_sum:
            max_sum = current_sum
            buy_day = temp_start
            sell_day = i

    return max_sum, buy_day + 1, sell_day + 1  # 天数从1开始

# 测试示例
profits = [3,2,1,-7,5,2,-1,3,-1]
max_profit, buy_day, sell_day = maxProfit(profits)
print(f"最大收益: {max_profit}, 买入日: 第{buy_day}天, 卖出日: 第{sell_day}天")
# your algorithm time complexity is: O(n)
````



### 问题3分析

* **时间复杂度**：

  * 遍历数组一次，O(n)

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1

idea:

````python
def findLocalMin(A):
    left, right = 0, len(A) - 1
    while left <= right:
        mid = (left + right) // 2
        if mid > 0 and mid < len(A) - 1:
            if A[mid] < A[mid-1] and A[mid] < A[mid+1]:
                return mid
            elif A[mid-1] < A[mid]:
                right = mid - 1
            else:
                left = mid + 1
        elif mid == 0:
            return 0 if A[0] < A[1] else 1
        else:  # mid == len(A)-1
            return len(A)-1 if A[-1] < A[-2] else len(A)-2

# 示例
A = [9, 3, 7, 2, 1, 4, 5]
idx = findLocalMin(A)
print(f"任意局部最小值: A[{idx}] = {A[idx]}")
# your algorithm time complexity is: O(log(n))
````

e.g.：

数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3

idea:

````python
arr = [5,4,3,2,6,1,88,33,22,107]
k = 3

# 排序法
sorted_arr = sorted(arr)
top_k = sorted_arr[:k]
print(f"排序法前{k}小数: {top_k}")
# your algorithm time complexity is: O(n log(n))
````

1.思路：先对数组排序，再输出前 k 个元素。

推荐使用 快速排序 或 内置 Timsort（Python sorted）
时间复杂度：O(n log n)
适用场景：数组不大，可以全部存入内存

````python
import heapq

arr = [5,4,3,2,6,1,88,33,22,107]
k = 3

# 构建最小堆
heap = []
for num in arr:
    heapq.heappush(heap, num)

# 取前k小
top_k = [heapq.heappop(heap) for _ in range(k)]
print(f"堆法前{k}小数: {top_k}")
# your algorithm time complexity is: O(n + k log(n))
````

思路：

将 n 个元素插入最小堆（min-heap）
调用 k 次 Extract-min 输出最小的 k 个元素
插入堆的时间复杂度：
直接插入 n 个元素：O(n log n)
每次 Extract-min：O(log n)，共 k 次 → O(k log n)
总时间复杂度：O(n log n + k log n) ≈ O(n log n)

````python
import heapq

arr = [5,4,3,2,6,1,88,33,22,107]
k = 3

# 最大堆存前k小
max_heap = []

for num in arr:
    if len(max_heap) < k:
        heapq.heappush(max_heap, -num)  # 用负号模拟最大堆
    else:
        if num < -max_heap[0]:
            heapq.heappop(max_heap)
            heapq.heappush(max_heap, -num)

# 输出按升序
top_k = sorted([-x for x in max_heap])
print(f"外部内存法前{k}小数: {top_k}")
# your algorithm time complexity is: O(n log(k))
````

思路：

假设数组太大不能全部加载到内存
可以用 大小为 k 的最大堆 保存当前最小的 k 个元素
遍历数据流：
堆未满 → 直接插入
堆满且当前元素 < 堆顶 → 弹出堆顶，插入当前元素
遍历结束后堆中就是前 k 小的元素
时间复杂度：O(n log k)
空间复杂度：O(k)

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]

idea:

````python
def planetCollision(A):
    stack = []
    for p in A:
        alive = True
        while alive and stack and stack[-1] > 0 and p < 0:
            top = stack[-1]
            # 栈顶小于当前，栈顶消失
            if abs(top) < abs(p):
                stack.pop()
            # 栈顶等于当前，右向左消失，左向右保留
            elif abs(top) == abs(p):
                alive = False  # 当前消失
                break
            # 栈顶更大，当前消失
            else:
                alive = False
                break
        if alive:
            stack.append(p)
    return stack

# 示例测试
A1 = [-3,-6,2,8,5,-8,9,-2,1]
A2 = [23,-8,9,-3,-7,9,-23,22]

res1 = planetCollision(A1)
res2 = planetCollision(A2)
res1, res2
````

### 算法分析

* **思路**：

  1. 使用栈保存左向右星球（正数）
  2. 遍历数组：

     * 栈顶 >0 且当前 <0 → 循环碰撞
     * 质量比较，质量小的消失，质量相等按题目规则处理
     * 当前存活则入栈
  3. 栈最终就是存活星球顺序
* **时间复杂度**：O(n)
* **空间复杂度**：O(n)（栈）




## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

idea:

````python
# add your code here
def knapsack_01(items, W):
    n = len(items)
    # dp[i][w] 表示前 i 件物品，容量为 w 的最大价值
    dp = [[0] * (W+1) for _ in range(n+1)]
    
    for i in range(1, n+1):
        weight, value = items[i-1]
        for w in range(W+1):
            if w >= weight:
                dp[i][w] = max(dp[i-1][w], dp[i-1][w-weight] + value)
            else:
                dp[i][w] = dp[i-1][w]
    
    return dp[n][W]

# 示例
items = [(5,10), (4,40), (6,30), (3,50)]
W = 9
max_value = knapsack_01(items, W)
print(f"最大价值: {max_value}")
# your algorithm time complexity is: O(n*W)
````





* **解题思路**：

  * 动态规划，状态 `dp[i][w]` 表示前 i 件物品、容量 w 的最大价值
  * 转移方程：
    [
    dp[i][w] =
    \begin{cases}
    \max(dp[i-1][w], dp[i-1][w-weight_i] + value_i), & w \ge weight_i \
    dp[i-1][w], & w < weight_i
    \end{cases}
    ]
  * 最终答案：`dp[n][W]`

* **时间复杂度**：O(n * W)

给你二叉搜索树的根节点 root ，该树中的恰好两个节点的值被错误地交换。请在不改变其结构的情况下，恢复这棵树 。设计算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现

e.g.:
---
1.

输入：root = [1,3,null,null,2]， 输出为：[3,1,null,null,2]， 

解释：3 不能是 1 的左孩子，因为 3 > 1 。交换 1 和 3 使二叉搜索树有效。

2.

 <div style="text-align:center;width:200px;"> <img alt="" src="./fig/hw3q9-1.jpg"> </div>
输入： root=[3,1,4,null,null,2]， 输出为：[2,1,4,null,null,3]

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。

idea:

````python
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def recoverTree(root):
    stack = []
    x = y = prev = None
    node = root
    
    # 中序遍历
    while stack or node:
        while node:
            stack.append(node)
            node = node.left
        node = stack.pop()
        if prev and node.val < prev.val:
            y = node
            if not x:
                x = prev
            else:
                break
        prev = node
        node = node.right
    
    # 交换两个错误节点的值
    if x and y:
        x.val, y.val = y.val, x.val

# 辅助函数：中序遍历打印
def inorder(root):
    return inorder(root.left) + [root.val] + inorder(root.right) if root else []

# 示例 1
root1 = TreeNode(1)
root1.left = TreeNode(3)
root1.left.right = TreeNode(2)
print("原 BST (中序):", inorder(root1))
recoverTree(root1)
print("恢复后 BST (中序):", inorder(root1))  # 输出应为 [1,2,3]

# 示例 2
root2 = TreeNode(3)
root2.left = TreeNode(1)
root2.right = TreeNode(4)
root2.right.left = TreeNode(2)
print("原 BST (中序):", inorder(root2))
recoverTree(root2)
print("恢复后 BST (中序):", inorder(root2))  # 输出应为 [1,2,3,4]
````



### 精简解题思路与时间复杂度分析

* **思路**：

  1. 使用中序遍历 BST，找到序列中**逆序对**。
  2. 第一个逆序对的前一个节点记为 `x`，后一个节点记为 `y`。
  3. 如果出现第二个逆序对，更新 `y`。
  4. 遍历完成后交换 `x.val` 和 `y.val` 即可恢复 BST。
* **时间复杂度**：O(n)，遍历每个节点一次
* **空间复杂度**：

  * 如果使用递归栈，最坏 O(h)（h 为树高度）
  * 如果迭代 + 显式栈，最坏 O(n)

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。

````python
# add your code here
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def sumNumbers(root):
    def dfs(node, current):
        if not node:
            return 0
        current = current * 10 + node.val
        if not node.left and not node.right:
            return current
        return dfs(node.left, current) + dfs(node.right, current)
    
    return dfs(root, 0)

# 示例 1
root1 = TreeNode(1)
root1.left = TreeNode(2)
root1.right = TreeNode(3)
print("示例 1 输出:", sumNumbers(root1))  # 输出 25

# 示例 2
root2 = TreeNode(4)
root2.left = TreeNode(9)
root2.right = TreeNode(0)
root2.left.left = TreeNode(5)
root2.left.right = TreeNode(1)
print("示例 2 输出:", sumNumbers(root2))  # 输出 1026
# your algorithm time complexity is: O(n)
````



* **解题思路**：

  * 使用 **深度优先搜索（DFS）** 遍历树
  * 对每条路径，将数字累加：
    [
    current = current \times 10 + node.val
    ]
  * 到达叶节点时，将当前路径数字加入总和
  * 遍历完成后返回总和

* **时间复杂度**：O(n)，遍历每个节点一次

* **空间复杂度**：O(h)，递归栈空间（h 为树高度）

* **功能**：

  * 返回从根到叶节点生成的所有数字之和
  * 示例：

    * `[1,2,3]` → 路径数字 12 + 13 = 25
    * `[4,9,0,5,1]` → 路径数字 495 + 491 + 40 = 1026
